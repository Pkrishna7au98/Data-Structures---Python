### Circular Linked list traversal and insertion of element at head position

```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
class CirLinList:
    
    def __init__(self):
        self.head = None
        
    def PushCLL(self, new_data):
        
        new_node = Node(new_data)
        temp = self.head
        
        new_node.next = self.head
        
        if self.head is not None:
            while (temp.next != self.head):
                temp = temp.next
            temp.next = new_node
        else:
            new_node.next = new_node
            
        self.head = new_node
        
    
    def Print_CLL(self):
        temp = self.head
        
        while temp is not None:
            print(temp.data, end = " ")
            temp = temp.next
            
            if temp == self.head:
                break
                

if __name__ == "__main__":
    
    Sin_CLL = CirLinList()
    
    Sin_CLL.PushCLL(11)
    Sin_CLL.PushCLL(12)
    Sin_CLL.PushCLL(13)
    Sin_CLL.PushCLL(14)
    Sin_CLL.PushCLL(15)
        
        
    Sin_CLL.Print_CLL()
            
```

### Sorted insert for circular linked list


```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
class CirLinList:
    
    def __init__(self):
        self.head = None
        
    def PushCLL(self, new_data):
        
        new_node = Node(new_data)
        temp = self.head
        
        new_node.next = self.head
        
        if self.head is not None:
            while (temp.next != self.head):
                temp = temp.next
            temp.next = new_node
        else:
            new_node.next = new_node
            
        self.head = new_node
    
    def sorted_insert(self,val):
        node_val = Node(val)
        
        if self.head.data > val:
            self.PushCLL(val)
            return
                
        ins = self.head
        while ins.next.data < val:
            ins = ins.next
            
            if ins.next == self.head:
                break
            
        if ins.next == self.head:
            ins.next = node_val
            node_val.next = self.head

            
        else:  
            temp = ins.next
            ins.next = node_val
            node_val.next = temp

    
    def Print_CLL(self):
        temp = self.head
        
        while temp is not None:
            print(temp.data, end = " ")
            temp = temp.next
            
            if temp == self.head:
                break
                

if __name__ == "__main__":
    
    Sin_CLL = CirLinList()
    
    Sin_CLL.PushCLL(11)
    Sin_CLL.PushCLL(8)
    Sin_CLL.PushCLL(5)
    Sin_CLL.PushCLL(3)
    Sin_CLL.PushCLL(1)
        
        
    Sin_CLL.Print_CLL()
    print()
    Sin_CLL.sorted_insert(15)
    Sin_CLL.sorted_insert(0)
    Sin_CLL.sorted_insert(6)
    Sin_CLL.Print_CLL()
```

### Circular Singly Linked List | Insertion


###### At the Head position, if the list is Null

```python
def Header(self, val):
    node = Node(val)
    self.head = node
    self.head.next = self.head
```

###### At the beginning/start of the list

```python
def HeadPush(self,val):
    node = Node(val)
    
    temp = self.head
    self.head = node
    node.next = self.head
```

###### At the end Position of the list

```python
def Appender(self, val):
    new_node = Node(val)
    
    last = self.head
    
    while last.next != self.head:
        last = last.next
        
    last.next = new_node
    new_node.next = self.head
```

###### At any position between head and last node

```python
def MiddleMan(self,prev_node, val):
    new_node = Node(val)
    
    temp = prev_node.next
    prev_node.next = new_node
    new_node.next = temp
    
```

### Deletion from a Circular Linked List


```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
class CirLinList:
    
    def __init__(self):
        self.head = None
        
    def PushCLL(self, new_data):
        
        new_node = Node(new_data)
        temp = self.head
        
        new_node.next = self.head
        
        if self.head is not None:
            while (temp.next != self.head):
                temp = temp.next
            temp.next = new_node
        else:
            new_node.next = new_node
            
        self.head = new_node
    
    def DeleteNode(self, data):
        
        if self.head is None:
            return
        
        if self.head.data == data and self.head.next == self.head:
            self.head = None
            return
        
        deleter = self.head
    
        while deleter.next.data != data:
            deleter = deleter.next
        
        if deleter.next == self.head:
            deleter.next = self.head.next
            self.head = self.head.next
        else:
            deleter.next = deleter.next.next
        
    
    def Print_CLL(self):
        temp = self.head
        
        while temp is not None:
            print(temp.data, end = " ")
            temp = temp.next
            
            if temp == self.head:
                break
                

if __name__ == "__main__":
    
    Sin_CLL = CirLinList()
    
    Sin_CLL.PushCLL(11)
    Sin_CLL.PushCLL(8)
    Sin_CLL.PushCLL(5)
    Sin_CLL.PushCLL(3)
    Sin_CLL.PushCLL(1)
        
    Sin_CLL.Print_CLL()
    print()
    Sin_CLL.DeleteNode(11)
    Sin_CLL.Print_CLL()
```

### Exchange first and last nodes in Circular Linked List


It can be done two ways, either swapping the values of both nodes, or swapping the nodes itself.

```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
class CirLinList:
    
    def __init__(self):
        self.head = None
        
    def PushCLL(self, new_data):
        
        new_node = Node(new_data)
        temp = self.head
        
        new_node.next = self.head
        
        if self.head is not None:
            while (temp.next != self.head):
                temp = temp.next
            temp.next = new_node
        else:
            new_node.next = new_node
            
        self.head = new_node
    
    def Exchanger(self):
        
        var = self.head
        
        while var.next != self.head:
            var = var.next
            
        self.head.data , var.data = var.data , self.head.data
    
    def Print_CLL(self):
        temp = self.head
        
        while temp is not None:
            print(temp.data, end = " ")
            temp = temp.next
            
            if temp == self.head:
                break
                

if __name__ == "__main__":
    
    Sin_CLL = CirLinList()
    
    Sin_CLL.PushCLL(9)
    Sin_CLL.PushCLL(8)
    Sin_CLL.PushCLL(7)
    Sin_CLL.PushCLL(6)
    Sin_CLL.PushCLL(5)
    Sin_CLL.PushCLL(4)
    Sin_CLL.PushCLL(3)
    Sin_CLL.PushCLL(2)
        
    Sin_CLL.Exchanger()
    Sin_CLL.Print_CLL()
 
```
