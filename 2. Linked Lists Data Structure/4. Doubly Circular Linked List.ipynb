### Doubly Circular Linked List Intro and Insertion 

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None
        
        
class DCLL_List:
    def __init__(self):
        self.head = None
    
    # If list is empty, then assigning a node which is circular, with next and prev 
    def InNull_DCLL(self, new_data):
        new_node = Node(new_data)
        
        new_node.next = new_node
        new_node.prev = new_node
        
        self.head = new_node
    
    #if List has some nodes and we want to add node at the end
    
    def Append_DCLL(self,appender):
        ap_node = Node(appender)
        
        vals = self.head
        
        while vals.next != self.head:
            vals = vals.next
            
        vals.next = ap_node
        ap_node.prev = vals
        ap_node.next = self.head
        self.head.prev = ap_node

        
    def Push_DCLL(self,pusher):
        Pnode = Node(pusher)
        
        last = self.head.prev
        
        last.next = Pnode
        Pnode.prev = last
        Pnode.next = self.head
        self.head.prev = Pnode
        self.head = Pnode
        
    
    def print_DCLL(self):
        
        elem = self.head
        while True:
            print(elem.data , end = " ")
            elem = elem.next
            
            if elem == self.head:
                print()
                break

if __name__ == "__main__":
    
    dcll = DCLL_List()
    
    dcll.InNull_DCLL(4)
    dcll.Append_DCLL(5)
    dcll.Append_DCLL(2)
    dcll.Append_DCLL(8)
    
    print("Original list is : ", end = " ")
    dcll.print_DCLL()
    
    dcll.Push_DCLL(1)
    print("After the above operation : ", end = " ")
    dcll.print_DCLL()
```

### Doubly Circular Linked List | Set 2 (Deletion)


```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None
        
        
class DCLL_List:
    def __init__(self):
        self.head = None
    
    # If list is empty, then assigning a node which is circular, with next and prev 
    def InNull_DCLL(self, new_data):
        new_node = Node(new_data)
        
        new_node.next = new_node
        new_node.prev = new_node
        
        self.head = new_node
    
    def Push_DCLL(self,pusher):
        Pnode = Node(pusher)
        
        last = self.head.prev
        
        last.next = Pnode
        Pnode.prev = last
        Pnode.next = self.head
        self.head.prev = Pnode
        self.head = Pnode
        
    def Delete_DCLL(self, val):
               
        if self.head is None:
            return
        
        if self.head.data == val and self.head.next == self.head:
            self.head = None
            return None
        
        else:
            hover = self.head
            
            while True: 
                
                if hover.data == val:
                    last = hover.prev

                    last.next = last.next.next
                    hover.next.prev = last
                    
                    if val == self.head.data:
                        self.head = last.next
                    
                    return 
                
                else:
                    hover = hover.next
                
                if hover == self.head:
                    return print("The given value is not found in the DCLL")
                
                
    def print_DCLL(self):
        if self.head is None:
            return
        
        elem = self.head
        while True:
            print(elem.data , end = " ")
            elem = elem.next
            
            if elem == self.head:
                print()
                break

if __name__ == "__main__":
    
    dcll = DCLL_List()
    
    dcll.InNull_DCLL(4)
    dcll.Push_DCLL(6)
    dcll.Push_DCLL(2)
    dcll.Push_DCLL(3)
    dcll.Push_DCLL(5)
    dcll.Push_DCLL(1)
    
    
    print("Original list is : ", end = " ")
    dcll.print_DCLL()
    
    print("After the above operation : ", end = " ")
    dcll.Delete_DCLL(2)
    dcll.Delete_DCLL(1)
    dcll.Delete_DCLL(4)
    dcll.print_DCLL()
```
