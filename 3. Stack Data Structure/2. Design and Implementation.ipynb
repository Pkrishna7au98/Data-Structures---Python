### Design a stack with operations on middle element

```python
''' Python3 Program to implement a stack
that supports findMiddle()
and deleteMiddle in O(1) time '''

class Node:
    
    def __init__(self,data):
        self.next = None
        self.prev = None
        self.data = data
        

class MyStack:
    
    def __init__(self):
        self.head = None
        self.mid = None
        self.count = 0
        
def createStack(self):
    ms = MyStack()
    self.count = 0
    return ms


        
def Push(ms, new_data):
    new_node = Node(new_data)
    
    new_node.prev = None
    new_node.next = ms.head
    
    ms.count += 1
    
    if ms.count == 1:
        ms.mid = new_node
        
    else:
        ms.head.prev = new_node
        
        if ms.count % 2 != 0:
            ms.mid = ms.mid.prev
            
    ms.head = new_node
    
def Pop(ms):
    
    if ms.count == 0:
        print("The stack is empty")
        return 
    
    head = ms.head
    temp = head.data
    ms.head = head.next
    
    if ms.head != None:
        ms.head.prev = None
        
    ms.count -= 1
    
    if ms.count % 2 == 0:
        ms.mid  = ms.mid.next
        
    return temp

def FindMid(self):
    if ms.count == 0:
        print("The stack is empty")
        return
    return ms.mid.data

if __name__ == '__main__':
 
    ms = MyStack()
    Push(ms, 11)
    Push(ms, 22)
    Push(ms, 33)
    Push(ms, 44)
    Push(ms, 55)
    Push(ms, 66)
    Push(ms, 77)
 
    print("Item popped is " +
          str(Pop(ms)))
    print("Item popped is " +
          str(Pop(ms)))
    print("Middle Element is " +
          str(FindMid(ms)))
        
```

### How to efficiently implement k stacks in a single array?

```python
class kstacks:
    
    def __init__(self,n,k):
        self.free = 0
        self.i = 0
        
        self.n = n
        self.k = k
        
        self.arr = [-float("inf")] * self.n
        self.top = [-1] * k
        
        self.next = [i+1 for i in range(n)]
        self.next[self.n-1] = -1
        
    def IsEmpty(self,sn):
        return self.top[sn] == -1
    
    def IsFull(self):
        return self.free == -1
    
    def Push_KS(self, val , sn):
        
        if self.IsFull():
            print("Stack Overflow")
            return 
        
        i = self.free
        self.free = self.next[i]
        self.next[i] = self.top[sn]
        self.top[sn] = i
        
        self.arr[i] = val
    
    def Pop_KS(self, sn):
        if self.IsEmpty(sn):
            return None

        top_of_stack = self.top[sn]
        self.top[sn] = self.next[self.top[sn]]
        self.next[top_of_stack] = self.free
        self.free = top_of_stack
  
        return self.arr[top_of_stack]

    def printstack(self, sn):
        top_index = self.top[sn]
        while (top_index != -1):
            print(self.arr[top_index])
            top_index = self.next[top_index]
    
    def print_arr(self):
        print(self.arr)
        
if __name__ == "__main__":
      
    # Create 3 stacks using an 
    # array of size 10.
    kstacks = kstacks(10,3)
  
    # Push some items onto stack number 2.
    kstacks.Push_KS(15, 2)
    kstacks.Push_KS(45, 2)
  
    # Push some items onto stack number 1.
    kstacks.Push_KS(17, 1)
    kstacks.Push_KS(49, 1)
    kstacks.Push_KS(39, 1)
  
    # Push some items onto stack number 0.
    kstacks.Push_KS(11, 0)
    kstacks.Push_KS(9, 0)
    kstacks.Push_KS(7, 0)

    kstacks.print_arr()
    
    #Pop out the elements of given stacks (sn)
    kstacks.Pop_KS(0)
    kstacks.Pop_KS(2)
    
    
    kstacks.printstack(0)
    print()
    kstacks.printstack(1)
    print()
    kstacks.printstack(2)
```

### How to create mergable stack?


```python
class Node:
    
    def __init__(self, data):
        self.next = None
        self.prev = None
        self.data = data
        
class StackList:
    
    def __init__(self):
        self.head = None
        self.tail = None
        
    
    def Push(self, new_data):
        
        new_Node = Node(new_data)
        
        if self.head == None:
            self.head = new_Node
            self.tail = new_Node
            
        else:
            new_Node.prev = self.tail
            self.tail.next = new_Node
            self.tail = new_Node
            
    def Pop(self):
        
        if self.head == None:
            return
        
        if self.head == self.tail:
            self.head = None
            self.tail = None
            
        else:
            temp = self.tail
            self.tail = self.tail.prev
            temp.prev = None
            
    def Merge_Stack(self, stack2):

        if stack2.head == None:
            return
        
        if self.head == None:
            self.head = stack2.head
            self.tail = stack2.tail
            return
        
        self.head.prev = stack2.tail
        stack2.tail.next = self.head
        self.head = stack2.head
    
    def DisplayStack(self):
         
        if (self.tail != None):
            n = self.tail
             
            while (n != None):
                print(n.data, end = " ")
                n = n.prev
 
            print()
 
        else:
            print("Stack Underflow")

if __name__ == "__main__":
    
    Mystack1 = StackList()
    Mystack2 = StackList()
    
    Mystack1.Push(1)
    Mystack1.Push(4)
    Mystack1.Push(3)
    Mystack1.Push(2)
    
    Mystack2.Push(6)
    Mystack2.Push(7)
    Mystack2.Push(8)
    Mystack2.Push(9)

    Mystack1.Pop()
    
    print("Stack 1 is : ", end = " ")
    Mystack1.DisplayStack()
    print()
    
    Mystack1.Merge_Stack(Mystack2)
    
    print("Stack 1 and Stack 2 are stacked together, The new stack is: ", end=" ")
    Mystack1.DisplayStack()
```
